```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
import numpy as np 

x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
#number of rows, columns, number of the plot you are refering 
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

# position of the axes[Left, Bottom, width, height]
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("LAGER PLOT")

axes2.plot(y,x)

axes2.set_title("SMALLER PLOT")
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
#small layout
# plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')

ax.plot(x, x**3, label='X Cubed')

# best 0 
ax.legend(loc='best')
```

# Plot Appearance


```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#linewidth = lw, linestyle = ls
ax.plot(x,y,color='purple', lw=1, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#linewidth = lw, linestyle = ls
ax.plot(x,y,color='purple', lw=1, alpha=1, ls='-', marker='o',
        markersize=10,markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') # RGB Hex code

ax.plot(x,y-3,color='green', lw=1, alpha=1, ls='-', marker='s',
        markersize=5,markerfacecolor='blue', markeredgewidth=3, markeredgecolor='black')
```
